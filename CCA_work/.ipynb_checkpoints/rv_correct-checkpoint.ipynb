```python
import matplotlib.pyplot as plt
from astropy import *
import astropy.units as u
from astropy.io import fits
import numpy as np
import glob as glob
import pandas as pd
from astropy.table import Table,vstack,hstack

import subprocess
import sys
import os
import logging
import numpy as np

from scipy.interpolate import interp1d
import time
from astropy.stats import sigma_clip
import scipy
from scipy.interpolate import interp1d
from scipy import interpolate

ispec_dir = '/uufs/astro.utah.edu/common/home/u1363702/iSpec'
sys.path.append(ispec_dir)


import ispec
def determine_radial_velocity_with_template(waves, fluxes):
    mu_cas_spectrum = Table([waves*.1, fluxes, np.repeat(0,len(fluxes))], names=('waveobs', 'flux', 'err'))
    #--- Radial Velocity determination with template -------------------------------
    #logging.info("Radial velocity determination with template...")
    # - Read synthetic template
    # template = ispec.read_spectrum(ispec_dir + "/input/spectra/templates/Atlas.Arcturus.372_926nm/template.txt.gz")
    template = ispec.read_spectrum(path + 'template.txt.gz')
    #template = ispec.read_spectrum(ispec_dir + "/input/spectra/templates/NARVAL.Sun.370_1048nm/template.txt.gz")
    #template = ispec.read_spectrum(ispec_dir + "/input/spectra/templates/Synth.Sun.300_1100nm/template.txt.gz")

    models, ccf = ispec.cross_correlate_with_template(mu_cas_spectrum, template, \
                            lower_velocity_limit=-200, upper_velocity_limit=200, \
                            velocity_step=1.0, fourier=False)

    # Number of models represent the number of components
    components = len(models)
    # First component:
    rv = np.round(models[0].mu(), 2) # km/s
    rv_err = np.round(models[0].emu(), 2) # km/s
    return rv, rv_err
```

```python
from scipy.interpolate import CubicSpline
```

```python
spectra_path = '/uufs/chpc.utah.edu/common/home/astro/zasowski/sinha/spectra/'
def get_spectra(obj_id,rv=-9999.9):
    wavelengths = []
    fluxes = []
    uncertainties = []
    # all_snrs = []
    for i in range(1,5):
        try:
            pathway = '/uufs/chpc.utah.edu/common/home/astro/zasowski/sinha/fits/galah/dr4/spectra/hermes/{}{}.fits'.format(obj_id,i)
            hdu = fits.open(pathway)
            flux = hdu[1].data
            err = hdu[2].data * flux

            # snr = hdu[0].header['SNR_AA']

            wave = []
            snrs = []
            for j in range(len(flux)):
                lamda = hdu[1].header['CRVAL1'] + hdu[1].header['CDELT1']*j
                wave.append(lamda)
                # snrs.append(snr)
            wave = np.array(wave)
            wavelengths.append(wave)
            fluxes.append(flux)
            uncertainties.append(err**2)
        except:
            print('no HDU {} for {}'.format(i,obj_id))

    if len(wavelengths) > 0:
        wavelength = np.concatenate(wavelengths)
        flux = np.concatenate(fluxes)
        uncertainty = np.concatenate(uncertainties)

        if rv == -9999.9:
            rv, rv_err = determine_radial_velocity_with_template(wavelength, flux)
        else:
            rv_err = 0
        corr_data = ispec.correct_velocity(Table([wavelength, flux, uncertainty], names=['waveobs','flux', 'err']), rv)

        waves = corr_data['waveobs']
        flux = corr_data['flux']
        flux_err = corr_data['err']
        flux_err[np.isfinite(flux_err)==False]= 1e-5

        start = 4712
        end = 7888
        resampled_waves = np.arange(start, end, 0.2)

        # data = np.dstack((waves,flux,flux_err))[0]
        cs_flux = CubicSpline(waves, flux)
        cs_err = CubicSpline(waves, flux_err)
        # cs_snr = CubicSpline(waves, snr_arr)
        
        
        resampled_flux = cs_flux(resampled_waves)
        resampled_flux_err = cs_err(resampled_waves)
        # resampled_snr = cs_snr(resampled_waves)

        resampled_flux[resampled_flux <=0]=np.nan
        resampled_flux[resampled_flux > 50]=np.nan

        resampled_flux_err[resampled_flux_err <=0]=np.nan
        resampled_flux_err[resampled_flux_err > 50]=np.nan

        data = np.dstack((resampled_waves,resampled_flux,resampled_flux_err))[0]

        star_id = '/galah_id_{}'.format(obj_id)

        filename = spectra_path+star_id + '.txt'
        np.savetxt(filename, data)
        return data, rv, rv_err
    else:
        return [0,0,0],0,0
```

```python
# cluster_members
ispec_rvs = []
ispec_rvs_errs = []

for i in cluster_members:
    # dr4_rvs.append(float(i['rv_comp_1']))
    # dr4_rvs_errs.append(float(i['e_rv_comp_1']))
    
    data, rv, rv_err = get_spectra(i['sobject_id'])
    ispec_rvs.append(rv)
    ispec_rvs_errs.append(rv_err)
    if rv > abs(50):
        data, rv, rv_err = get_spectra(i['sobject_id'],rv=float(i['rv_comp_1']))

cluster_members['ispec_rv'] = ispec_rvs
cluster_members['e_ispec_rv'] = ispec_rvs_errs
```
