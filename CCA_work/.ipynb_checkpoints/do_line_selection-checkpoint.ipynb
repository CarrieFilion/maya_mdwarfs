```python
teff_range = np.arange(3000,4250,250)
logg_range = np.arange(4, 4.8, 0.2)
m_h_range = np.arange(-0.5, 0.25,0.25)

from itertools import product
# Generate all combinations
combinations = list(product(teff_range, np.round(logg_range,1)))
combinations = list(product(combinations, np.round(m_h_range,2)))
```

```python
from juliacall import Main as jl
jl.seval("using Korg")
Korg = jl.Korg

lines = Korg.get_GES_linelist()
waves = Korg.Wavelengths(np.arange(4713, 7887.2, 0.2))
```

```python
teff = combinations[0][0][0]
logg = combinations[0][0][1]
m_h = combinations[0][1]
vmicro = np.round(2.488 - 0.8665*logg+ 0.1567*logg**2,3)
```

```python
def generate_synthetic_spectra(teff, logg, vmicro, m_h):
    abunds = Korg.format_A_X(m_h)
    atmosphere =  Korg.interpolate_marcs(teff, logg, abunds)
    spectrum = Korg.synthesize(atmosphere, lines,abunds, waves,vmic=vmicro)
    norm_flux = np.array(spectrum.flux) / np.array(spectrum.cntm)
    norm_flux = np.array(Korg.apply_LSF(norm_flux, waves, 28_000))
    norm_flux = Korg.apply_rotation(norm_flux, waves, 5)
    return norm_flux
```

```python
synthetic_flux = generate_synthetic_spectra(teff, logg, vmicro, m_h)
```
